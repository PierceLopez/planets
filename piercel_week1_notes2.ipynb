Dictionaries store _key:value_ pairs.

```

dictionary = {"key holding a numeric":1.23, "key holding a string":"string}

Recall lists

```
# sample list
list = [value1, value2, value3]

# deleting a key:value pair

### Section 4: Pandas (part 2)

Ways of selecting columns:
1. Single bracket notation
  * outputs a Series (1D array)
2. Double bracket notation
  * outputs an n-column DataFrame depending on how many columns are called

**Note:** We can do slicing if we want to select multiple continuous rows with ease!
```
# single bracket notation
df[column_name]
df[row_index1:row_index2]

# double bracket notation
df[[column_name]]
df[[column_name1, column_name2]]
```

Ways of selecting rows:
1. _loc[]_
  * outputs a Series (1D array)
2. _loc[[]]_
  * outputs an n-row DataFrame depending on how many rows are called
3. _iloc[]_
  * outputs a Series (1D array)
4. _iloc[[]]_
  * outputs an n-row DataFrame depending on how many rows are called

```
# single bracket loc
df.loc[row_name]

# double bracket loc
df.loc[[row_name1]]
df.loc[[row_name1, row_name2]]

# single bracket iloc
df.iloc[row_index]

# double bracket iloc
df.iloc[[row_index]]
```

**Note:** Python starts indexing at 0!

We can obtain a part of the DataFrame by specifying rows and columns at the same time.

```
# obtain certain rows and columns of the DataFrame
df.loc[[row_name1, row_name2, row_name3], [column_name1, column_name2]]
df.iloc[[row_index1, row_index2, row_index3], [column_index1, column_index2]]

# obtain slices (adjacent rows and columns) using ':'
df.loc[row_name1:row_name2, column_name1:column_name2]
df.loc[[row_name1], column_name1:column_name2]
df.iloc[row_index1:row_index2, column_index1:column_index2]
df.iloc[row_index1:row_index2, [column_index1]]
```

We can take all rows or columns using ':' alone!

```
# obtain all rows
df.loc[:, [column_name1, column_name2]]
df.loc[:, column_name1:column_name2]
df.iloc[:,[column_index1, column_index2]]
df.iloc[:,column_index1:column_index2]

# obtain all columns
df.loc[[row_name1, row_name2], :]
df.loc[row_name1:row_name2, :]
df.iloc[[row_index1, row_index2], :]
df.iloc[row_index1:row_index2, :]
```

**Note:** When indexing using colons, you do not need to to place them inside brackets in a double bracket notation.

## Chapter 4: Loops

### Section 1: while loop

<br>

`while` loops will always run until the condition becomes false.

```
# while loop syntax

while condition :
    expression
```

<br>

### Section 2: for loop

`for` loops will always run until every item in the collection or sequence is iterated on.

```
# for loop syntax

for item in collection :
    expression

# for loop syntax that includes the collection indices

for index, item in enumerate(collection) :
    print(str(index) + ":" + item)
```
**Note:** Some examples of colletions or sequences we can loop on are strings, lists, arrays, and number ranges.

<br>

### Section 3: Loop data structures (part 1)

In the case of...
* dictionaries, we just modify the _collection_ component of the for loop by adding the _items()_ method (this call the elements inside the dictionary).
* 2D NumPy arrays, we just use the _np.nditer()_ function on our collection to iterate on every element on the 2D array.

```
# looping over a dictionary

for key, value in dictionary.items() :
    expression

# looping over a 2D NumPy array

for item in np.nditer(twoDarray) :
    expression
```

**Note:** Order matters!
<br>

### Section 4: Loop data structures (part 2)

In the case of pandas DataFrames we use the _iterrows()_ method to obtain the row labels and values.

```
# looping over a pandas DataFrame

for row_label, row_value in df.iterrows() :
    print(row_label)
    print(row_value)

# looping over a pandas DataFrame and secleting only a specific column

for row_label, row_value in df.iterrows() :
    print(str(row_label)+ ":" + row_value["column_name"]
```

Using the `for` loop, we can also add a new column in the DataFrame!

```
# adding a column (length of a particular string column named 'country' for example) using a for loop 

for row_label, row_value in df.iterrows() :
    df.loc[row_label, "new_column_name"] = len(row_value["country"])
```

A simpler and more effecient way to do this is by using the _apply()_ method, which applies a certain function on a selected array. 

```
# adding a column using apply()
df["new_column_name"] = df["country"].apply(len)
```

### Section 2: Random Walk

Random walks are successions of random steps. We take coin tossing as a example (10 coin tosses) to generate two different outputs:

```
# first output (printing outcome of every toss)
# second output (incrementation is applied when an iteration outputs tails)

# initialization
outcome1 = []
outcome2 = [0]

for toss in range(10) :
    face = np.random.randint(0, 2)
    if face == 1 :
      outcome1.append('tails')
    else :
      outcome1.append('heads')
    outcome2.append(outcome2[toss] + face)
print(outcome1) # first output
print(outcome2) # second output

# Non-indented lines signify that they are not part of the loop anymore.
```

Notice that the first output is just a collection of random steps, while the second output made a random walk basing from the collection of random steps from the first output!

<br>

### Section 3: Distribution

Let us consider making a distribution that collects the number of times 'tails' appeared in a 10-coin-toss run 100 times:

```
# initialization
final_outcomes = []

for run in range(100) :
    outcome2 = [0]
    for toss in range(10) :
        face = np.random.randint(0, 2)
        outcome2.append(outcome2[toss] + face)
    final_outcomes.append(outcome2[-1])  
print(final_outcomes)
```

We can visualize `final_outcomes` using a histogram:

```
plt.hist(final_outcomes, bins = 10)
plt.show()
```
